```python
u = []
i = []
r = []
for user in db.collection('users').get():
    uid = user.id
    for img in user.reference.collection('ratings').get():
        iid = img.id
        rating = img.to_dict()['rating']
        
        u.append(uid)
        i.append(iid)
        r.append(rating)
```

```python
ratings = pd.DataFrame({'userId':u, 'imageId':i, 'rating':r})
```
